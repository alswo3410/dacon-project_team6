```python
import pandas as pd
```

- 지가지수 데이터 불러오기

```python
df = pd.read_csv('raw/지가지수/(연) 지역별 지가지수.csv', encoding = 'cp949')
```

```python
df = df.loc[2:].reset_index(drop = True)
df.drop('No', axis = 1, inplace = True)
```

```python
df
```

```python
df = df.melt(id_vars = '지역', value_vars = df.columns[1:])
df.columns = ['시도', '연도', '지가지수']
```

```python
# 연도에서 '년'을 제거한 뒤 정수 타입으로 변경
df['연도'] = df['연도'].str.replace('년', '').astype(int)
```

```python
df
```

```python
df.to_csv('전처리/지가지수_전처리.csv', index = False)
```
