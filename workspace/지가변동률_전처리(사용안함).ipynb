```python
import pandas as pd
```

```python
df1 = pd.read_csv('전처리/주택소유율에따른_경제안정성과고용률.csv')
df2 = pd.read_excel('raw/지가변동률/시군구별_연간_지가변동률_통계(1975~2018년).xlsx')
```

```python
df2.columns
```

```python
# 공백 제거
df2['지역'] = df2['지역'].str.replace(' ', '')
```

```python
df2.loc[df2['지역'] == '서울특별시']
```

```python
# 시도 리스트
sido_list = ['서울특별시', '부산광역시', '대구광역시', '인천광역시', '광주광역시', '대전광역시',
             '울산광역시', '세종특별자치시', '경기도', '강원도', '충청북도', '충청남도',
             '전라북도', '전라남도', '경상북도', '경상남도', '제주특별자치도']
```

```python
# 2005년도 부터 사용
df2 = pd.concat([df2.loc[:, '지역'], df2.loc[:, '2005년':]], axis = 1)

# 시도 데이터만 사용
df2 = df2.loc[df2['지역'].isin(sido_list)].reset_index(drop = True)
```

```python
df2
```

```python
df2.columns[1:]
```

```python
df2 = df2.melt(id_vars = '지역', value_vars = df2.columns[1:])
df2.columns = ['시도', '연도', '지가변동률']
```

```python
# 세종시는 지가변동률 데이터가 없다.
df2.loc[df2['시도'] == '세종특별자치시']
```
